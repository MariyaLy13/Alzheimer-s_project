## Findings and conclusions
- The analysed dataset contains information on 760 patients with Alzheimer's disease (35.4%) and 1389 patients without Alzheimer's disease (64.6%).
- 50.8% of patients with Alzheimer's disease are male and 49.2% are female.
- The largest group of patients have Caucasian ethnicity - 1278 people (32.2% diagnosed with Alzheimer's disease), but Asians have the highest percentage of diagnosed Alzheimer's disease - 40.8%.
- The highest percentage of people diagnosed with Alzheimer's disease have a high school education.
- The analysis doesn't show any difference between most of the individual lifestyle factors in the group of patients with and without Alzheimer's disease, except for sleep quality. The percentage of patients diagnosed with Alzheimer's disease who sleep 4-5 hours is 5.5% higher than in patients without a diagnosis (35.5% to 30%).
- There wasn't a direct correlation between medical history and Alzheimer's diagnosis.
- Patients without Alzheimer's disease have lower HDL cholesterol levels.
- Patients with Alzheimer's disease are 3 times more likely to have memory complaints and behavioural problems.
- There are 10% more memory problems than behavioural problems in patients with Alzheimer's disease. Memory problems do not always lead to behavioural problems. Asians have fewer behavioural problems (21.4%) than other ethnic groups, even when memory problems are highest (42.9%).
Caucasians have a percentage of behavioural problems close to that of African Americans (27.6% to 28.1%) when the percentage of memory problems is about 5% lower (35.2% to 40.4%). These differences could be explored further to understand what helps some patients avoid behavioural problems when memory problems occur.
- The most common symptom experienced by people with Alzheimer's disease is forgetfulness (30.1%). Confusion (19.5%), difficulty doing tasks (16.3%), disorientation and personality changes (14.6% and 14.1% respectively).
- Forgetfulness is the most common symptom across all ethnic groups, but varies widely:
    - African Americans - 34.9%.
    - Caucasians - 30.9%.
    - Asians - 25.0%.
    - Other - 20.9%.
- African Americans report higher rates of confusion (21.9%) than others. Asians have the highest rate of difficulty in completing tasks (23.8%).

**1. Mapping the numeric values to their respective definitions for columns:**

**2. Convert PatientID to string**

```python
df['PatientID'] = df['PatientID'].astype(str)
```

**3. Checking for outliers using the IQR method**

```python
numeric_cols = [col for col in df.select_dtypes(include=['number']).columns if df[col].nunique() > 2]

def detect_outliers(df):
    outliers = {}
    
    for column in numeric_cols:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers[column] = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column].tolist()

    return outliers

outliers_numeric_cols = detect_outliers(df)

print("Outliers in numeric columns (excluding binary columns):")
for column, outliers in outliers_numeric_cols.items():
    print(f"{column}: {outliers}")
```

**5. What is the percentage of people diagnosed with Alzheimer's in each age group by ethnicity?**

```python
mosaic_data = df[df['Diagnosis'] == 1].groupby(['Ethnicity', 'Age_group']).size().unstack(fill_value=0)
mosaic_data = mosaic_data.div(mosaic_data.sum(axis=1), axis=0) * 100

mosaic_dict = {
    (ethnicity, age_group): value
    for ethnicity in mosaic_data.index
    for age_group, value in mosaic_data.loc[ethnicity].items()
}

fig, ax = plt.subplots(figsize=(8, 5))
props = lambda key: {'alpha': 0.8}
mosaic(mosaic_dict, properties=props, labelizer=lambda key: f"{mosaic_dict[key]:.1f}%", ax=ax)

plt.title("Proportion of Positive Diagnoses by Ethnicity and Age Group")
plt.ylabel("Proportion (%)")

plt.tight_layout()
plt.show()
```

**Conclusion:**
- The proportion of people with Alzheimer's disease is almost evenly distributed across all ethnic groups, except for Asians, where the proportion of people aged 70-79 with the disease is significantly lower than in other age groups. At the same time, Asians have the highest proportion of people aged 60-69 with Alzheimer's disease of any ethnic group.
- Asians have the highest proportion of people with Alzheimer's disease and develop the disease at a slightly earlier age.

**Perform a Mann-Whitney U test to understand if there is a difference between each of the life factors in patients with and without Alzheimer's disease.**

```python
life_factors_columns = ['BMI', 'Smoking', 'AlcoholConsumption', 'PhysicalActivity', 'DietQuality', 'SleepQuality']

diagnosis_0 = df[df['Diagnosis'] == 0]
diagnosis_1 = df[df['Diagnosis'] == 1]

results = {}

for column in life_factors_columns:
    data_0 = df[df['Diagnosis'] == 0][column]
    data_1 = df[df['Diagnosis'] == 1][column]
    
    stat, p_value = stats.mannwhitneyu(data_0, data_1, alternative='two-sided')
    
    results[column] = {'U-statistic': stat, 'p-value': p_value}

for column, stats in results.items():
    print(f"{column}: U-statistic = {stats['U-statistic']:.2f}, p-value = {stats['p-value']:.4f}")
    if stats['p-value'] < 0.05:
        print(f"  -> Significant difference between Diagnosis 0 and Diagnosis 1 for {column}")
    else:
        print(f"  -> No significant difference between Diagnosis 0 and Diagnosis 1 for {column}")
```

- U-statistic = 563558.00, p-value = 0.0094
- Interpretation: There is a significant difference in sleep quality between patients without Alzheimer's disease (diagnosis == 0) and patients with Alzheimer's disease (diagnosis == 1).
- **Conclusion:** Sleep quality is likely to be a factor that differs between patients with and without the condition represented by the diagnosis. Further research is needed to investigate its role.

### Medical History impact

- **14. What % of people diagnosed with Alzheimer's disease have diabetes?**
- **15. What % of people diagnosed with Alzheimer's have a family history of Alzheimer's?**
- **16. What % of people diagnosed with Alzheimer's have cardiovascular disease?**
- **17. What % of people diagnosed with Alzheimer's have a head injury?**
- **18. What % of people diagnosed with Alzheimer's disease have depression?**
- **19. What % of people diagnosed with Alzheimer's have hypertension?**

```python
medical_history_columns = ['Diabetes', 'FamilyHistoryAlzheimers', 'CardiovascularDisease', 
                           'HeadInjury', 'Depression', 'Hypertension']
medical_proportion_results = []

for column in medical_history_columns:
    counts = df.groupby(['Diagnosis', column]).size().unstack(fill_value=0)
    proportions = counts[1] / counts.sum(axis=1)
    medical_proportion_results.append(proportions)

medical_final_results = pd.concat(medical_proportion_results, keys=medical_history_columns, axis=1)

print("Proportions of Medical History Impact by Diagnosis (0 - Negative, 1 - Positive):")
print(medical_final_results.T)
```

```python
plt.figure(figsize=(14, 12))

for idx, column in enumerate(medical_history_columns):
    data = medical_final_results[column]
    
    ax = plt.subplot(2, 3, idx + 1)
    bars = data.plot(kind='bar', ax=ax, color=['lightblue', 'lightgreen'])
    
    ax.set_title(f'{column} by Diagnosis')
    ax.set_ylabel('Proportion')
    ax.set_xlabel('Alzheimer\'s Diagnosis')
    ax.set_xticklabels(['Negative (0)', 'Positive (1)'], rotation=0)
    ax.set_ylim(bottom=0.08)

    for bar in bars.patches:
        height = bar.get_height()
        percentage = f'{height * 100:.1f}%'
        ax.text(bar.get_x() + bar.get_width() / 2, height + 0.001, percentage, 
                ha='center', va='bottom', fontsize=10, color='black')
        
plt.tight_layout()
plt.show()
```

**Conclusion.**
- Patients with Alzheimer's disease are more likely to have hypertension or cardiovascular disease than patients without Alzheimer's disease.
- I will do a T-test to understand if the difference is significant.

**T-test**

```python
for column in medical_history_columns:
    group_0 = df[df['Diagnosis'] == 0][column]
    group_1 = df[df['Diagnosis'] == 1][column]
    t_stat, p_value = stats.ttest_ind(group_0, group_1)
    print(f"T-test for {column}: t-statistic = {t_stat:.2f}, p-value = {p_value:.3f}")
```

**Conclusion:** p-value > 0.05. No significant association between medical history columns and Alzheimer's diagnosis.
